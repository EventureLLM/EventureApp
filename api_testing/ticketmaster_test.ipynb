## Ticketmaster Test
Goal: Test ticketmaster API and learn how it works

```python
import requests
from dotenv import load_dotenv
import os

# Load environment variables from .env file
load_dotenv()

# Access the API key
ticketmaster_key = os.getenv("TICKETMASTER_API_KEY")
```

```python

def makeTicketmasterEventRequest(package, version, eventparams):
    # Format for looking at events: https://app.ticketmaster.com/{package}/{version}/events.json?{eventparams}apikey={API key}
    # eventparams needs to be separated by & symbols for each query. Check ticketmaster discovery page to see all query options

    url = f'https://app.ticketmaster.com/{package}/{version}/events.json?{eventparams}apikey={ticketmaster_key}'

    r = requests.get(url)

    return r 
```

```python
# Be sure to include an '&' at the end of your parameters for your query for your event for the request to pass through
r = makeTicketmasterEventRequest(package='discovery', version='v2', eventparams='stateCode=MI&')
r.status_code
```

```python
import json
r_json = r.json()
type(r_json)
# Testing to see what keys
print(r_json.keys())

# printing all events happening in michigan according to ticketmaster 
for i in range(len(r_json['_embedded']['events'])):
    print(r_json['_embedded']['events'][i]['name'])
```

## Tripadvisor Test
Goal: Test tripadvisor API and learn how it works

```python
import requests
from dotenv import load_dotenv
import os

# Load environment variables from .env file
load_dotenv()

# Access the API key
tripadvisor_key = os.getenv("TRIPADVISOR_API_KEY")
```

```python
# Might be interested in making a "Nearby search" request with an address for things nearby
# Might be interested in making a "Find search"

def makeTripadvisorFindRequest(Name_of_location):
    # replace spaces with '%' to be able to pass in the name of a location
    Name_of_location = Name_of_location.replace(" ", "%")

    print(Name_of_location) # Debugging

    url = f'https://api.content.tripadvisor.com/api/v1/location/search?key={tripadvisor_key}&searchQuery={Name_of_location}&radius=20&language=en'
    # url = f"https://api.content.tripadvisor.com/api/v1/location/search?key={tripadvisor_key}&searchQuery=Ann%20Arbor&radius=20&language=en"
    r = requests.get(url)
    return r 
```

Note: Searching for Ann Arbor isn't as simple as writing 'Ann Arbor'. It requires a '20' --> 'Ann 20Arbor'.

```python
r = makeTripadvisorFindRequest('Detroit')
r.status_code
```
